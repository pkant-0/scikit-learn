# write a python script to fit a model to the given data and
Predict the value of y at x = 50.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

# Using 'scr-dataset.csv'.

Uploaded this to the colab and using csv to further process. i.e loading, model fitting and predicting the value.

```python
# Loading Data
data = pd.read_csv('scr-dataset.csv')
x = data['x'].values.reshape(-1, 1)
y = data['y'].values
```

```python
# Fitting a linear model to the data

model = LinearRegression(fit_intercept=True)
model.fit(x, y)
```

```python
# Predict the value of y at x = 50

x_updated = np.array([[50]])
y_predict = model.predict(x_updated)

print("Predicted value of y at x = 50:", y_predict)
```

```python
plt.scatter(x, y, color='grey', label= 'Data points')
plt.plot(x, model.predict(x), color='red', label = "Fit line")
plt.scatter(x_updated, y_predict, color='green', label = "Predicted value")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')
plt.show()
```
